```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt    #vizualisation library (adv)
import seaborn as sb             #vizualisation library

%matplotlib inline
```

```python
pokemon=pd.read_csv('pokemon.csv')
print(pokemon.shape)
pokemon.head()
```

```python
sb.countplot(data=pokemon,x= 'generation_id')  #sb is seaborn
```

```python
base_color=sb.color_palette()[0]
sb.countplot(data=pokemon, x='generation_id', color=base_color)
```

```python
base_color = sb.color_palette()[1]
gen_order = pokemon['generation_id'].value_counts().index
sb.countplot(data = pokemon, x = 'generation_id', color = base_color, 
              order = gen_order)
```

```python
base_color = sb.color_palette()[2]
sb.countplot(data = pokemon, x = 'type_1', color = base_color)
```

```python
base_color = sb.color_palette()[2]
sb.countplot(data = pokemon, x = 'type_1', color = base_color)
plt.xticks(rotation=90);   #xticks - The ticks/labels present in x axis
```

```python
base_color = sb.color_palette()[2]
sb.countplot(data = pokemon, y = 'type_1', color = base_color)
```

```python
#counting missing data

pokemon.isna().sum()
```

```python
na_counts = pokemon.isna().sum()
base_color = sb.color_palette()[0]
sb.barplot(na_counts.index.values, na_counts, color = base_color)
plt.xticks(rotation=90);
```

```python
#piechart

sorted_counts = pokemon['generation_id'].value_counts()
plt.pie(sorted_counts,labels=sorted_counts.index, startangle=90, counterclock=False);
plt.axis('square');
```

```python
sorted_counts = pokemon['generation_id'].value_counts()
plt.pie(sorted_counts,labels=sorted_counts.index, startangle=90, counterclock=False , wedgeprops= {'width':0.4} );
plt.axis('square');
```

```python
sb.distplot(pokemon['speed'])
```

```python
sb.distplot(pokemon['speed'], kde=False)
```

```python
bins_edges= np.arange(0, pokemon['speed'].max() +1 , 5)  #to include max element use +1
sb.distplot(pokemon['speed'],bins=bins_edges, kde=False);  #alpha increase color contrast
```

```python
bins_edges= np.arange(0, pokemon['speed'].max() +1 , 5)  #to include max element use +1
sb.distplot(pokemon['speed'],bins=bins_edges, kde=False, hist_kws={'alpha' : 1});  #alpha increase color contrast
```
